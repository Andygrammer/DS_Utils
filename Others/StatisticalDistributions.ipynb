# 1. Import Libraries

```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

# 2. Normality Test

```python
# generate a sample with ~normal data (bell curve)
data = norm.rvs(size = 1000)

# plot the data
plt.hist(data, bins = 20) # 20 bars
plt.title('Normal data')
```

```python
# plot a probability normal distribution graph
fig, ax = plt.subplots()
stats.probplot(data, fit=True, plot=ax)
plt.show()
```

```python
# Shapiro test to check if data is a normal distribution

# if p > 0.05, it is impossible to reject null hypothesis therefore there is a normal distribution
stats.shapiro(data)
```

```python
# compare with a non normal distribution
data2 = skewnorm.rvs(4, size = 1000)

# plot the data
plt.hist(data2, bins = 20) # 20 bars
plt.title('Non normal data')
```

```python
# plot a probability for non normal distribution data
fig, ax = plt.subplots()
stats.probplot(data2, fit=True, plot=ax)
plt.show()
```

```python
# Shapiro test for non normal distribution
stats.shapiro(data2)
```
