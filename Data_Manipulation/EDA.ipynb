# 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

# 2. Prepare Data

```python
# Load the data
data = pd.read_csv('../datasets/melb_data.csv')
data.head()
```

# 3. EDA - Shape Exploration

```python
# 1. #rows vs. #columns
data.shape
```

```python
# 2. data types
data.dtypes
```

```python
# 3. display a few rows
data.head()
data.tail()
data.sample(5)
```

```python
# 4. plot each numeric feature ----- *********userful to detect outliers*********
data.hist(figsize=(14,14), xrot=45)
plt.show()
```

```python
# 5. summary statistics of the numerical features
data.describe()
```

```python
# 6. summary statistics of the categorical features
data.describe(include='object')
```

```python
# 7. plot each categorical feature ----- ******userful to detect sparse classes and mistakes in labeling******
for column in data.select_dtypes(include='object'):
    if data[column].nunique() < 10:     # 10 or fewer classes
        sns.countplot(y=column, data=data)
        plt.show()
```

```python
# 8. segment the target variable by categorical features  

# *** useful to determine which classes most affect the target variables ***

for column in data.select_dtypes(include='object'):
    if data[column].nunique() < 10:
        sns.boxplot(y=column, x='Price', data=data)
        plt.show()
```

```python
# 9. group numeric features by each categorical feature
for column in data.select_dtypes(include='object'):
    if data[column].nunique() < 10: # for only the 3 categorical features with less than 10 classes
        display(data.groupby(column).mean())
```

```python
# 10. correlations matrix for the different numerical features
"""""
A positive correlation means that as one feature increases the other one also increases, 
while a negative correlation means one feature increases as the other decreases. 
Correlations close to 0 indicate a weak relationship while closer to -1 or 1 signifies a strong relationship.
"""""
corrs = data.corr()
corrs
```

```python
# 11. heatmap of the correlations
"""
The cmap=â€˜RdBu_râ€™ argument tells the heatmap what colour palette to use.
A high positive correlation appears as dark red and a high negative correlation as dark blue.
Closer to white signifies a weak relationship. Read this nice tutorial for other color palettes.
annot=True includes the values of the correlations in the boxes for easier reading and interpretation.
"""
plt.figure(figsize=(10,8))
sns.heatmap(corrs, cmap='RdBu_r', annot=True)
plt.show()
```
